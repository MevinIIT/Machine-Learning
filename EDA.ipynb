```python
# ==========================================================
# Machine Learning Coursework
# Exploratory Data Analysis (EDA)
# Bank Marketing Dataset
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# LOAD DATASET
# ==========================================================

print("=" * 60)
print("LOAD DATASET")
print("=" * 60)

df = pd.read_csv("bank-full.csv", sep=";")

print(df.head())


# ==========================================================
# DATASET SHAPE
# ==========================================================

print("\n" + "=" * 60)
print("DATASET SHAPE")
print("=" * 60)

print("Rows    :", df.shape[0])
print("Columns :", df.shape[1])


# ==========================================================
# DATA TYPES
# ==========================================================

print("\n" + "=" * 60)
print("DATA TYPES")
print("=" * 60)

print(df.dtypes)


# ==========================================================
# MISSING VALUES
# ==========================================================

print("\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)

print(df.isnull().sum())


# ==========================================================
# DUPLICATE ROWS
# ==========================================================

print("\n" + "=" * 60)
print("DUPLICATE ROWS")
print("=" * 60)

print("Number of duplicate rows:", df.duplicated().sum())


# ==========================================================
# SUMMARY STATISTICS
# ==========================================================

print("\n" + "=" * 60)
print("SUMMARY STATISTICS")
print("=" * 60)

print(df.describe())
```
